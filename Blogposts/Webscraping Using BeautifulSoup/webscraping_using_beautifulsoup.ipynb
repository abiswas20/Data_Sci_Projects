We can follow the steps above repeatedly to get to the desired text. But there are more efficient ways which will be explored below.  
The first thing we'll look at is **finding all instances of a tag at once:**

```python
soup=BeautifulSoup(page.content,'html.parser')
soup.find_all('tr')
```

```python
#since find_all returns a list we can use list indexing to extract text
print(soup.find_all('tr')[0].get_text())
```

```python
print(soup.find_all('tr')[1].get_text())
```

```python
print(soup.find_all('tr')[2].get_text())
```

```python
for i in range(1,len(soup.find_all('tr'))):
    print(soup.find_all('tr')[i].get_text())
```

Now that we have seem how to access individual values, we are ready to put together all the data on the static website as a dataframe.

```python
soup.find_all('tr')
```

```python
import itertools

allKeys=['Name','Category','Calories','Fat','Carbs']

b=[]
menuList=[]
for i in range(1,len(soup.find_all('tr'))):
    a=soup.find_all('tr')[i].get_text().split('\n')[1:6]
    for element in a:
        menuList.append(element)
g=iter(menuList)
for j in range(1,len(soup.find_all('tr'))):
    dishDict=dict(zip(allKeys,g))
    b.append(dishDict)

pd.DataFrame(b)
```
